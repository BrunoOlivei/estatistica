```python

import numpy as np

data = [12, 34, 57, 29, 41, 67, 48, 33, 55, 38, 49, 26, 36, 28, 51]

stems = [int(x / 10) for x in data]
leaves = [x % 10 for x in data]

plt.stem(stems, leaves)
plt.xlim(min(stems)-1, max(stems)+1)
plt.title("Stem-and-Leaf Plot")
plt.xlabel("Stems")
plt.ylabel("Leaves")

plt.show()
```
